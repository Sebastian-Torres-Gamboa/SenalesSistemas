Representaciones de Fourier para Análisis de Señales en el Dominio de la Frecuencia
Serie de Fourier (Exponencial, Trigonométrica y Compacta)
Tipo de señales: Analiza señales periódicas en tiempo continuo.

Espectro: Discreto, compuesto por múltiplos de la frecuencia fundamental de la señal.

Semejanzas entre formas:

La forma exponencial es la base, mientras que las formas trigonométrica (senos/cosenos) y compacta (amplitud/fase) se derivan de ella.

Diferencias:

La trigonométrica emplea funciones sinusoidales reales.

La exponencial utiliza exponenciales complejas.

La compacta combina amplitud y fase en una expresión simplificada.

Transformada de Fourier
Tipo de señales: Diseñada para señales no periódicas en tiempo continuo.

Espectro: Continuo, cubriendo todas las frecuencias posibles.

Relación con la Serie de Fourier:

Es una generalización de la Serie de Fourier para señales no periódicas. Si el período de una señal periódica tiende a infinito, su Serie de Fourier converge a la Transformada de Fourier.

Diferencia clave:

Mientras la Serie de Fourier maneja señales periódicas con espectro discreto, la Transformada de Fourier trabaja con señales no periódicas y genera un espectro continuo.

Transformada Discreta de Fourier (DFT)
Tipo de señales: Usada para señales periódicas en tiempo discreto (o secuencias finitas tratadas como periódicas).

Espectro: Discreto, calculado en frecuencias específicas.

Relación con la DTFT:

La DFT es una versión muestreada de la DTFT, evaluada en frecuencias discretas. Es esencial en el procesamiento digital de señales.

Diferencia principal:

La DFT opera sobre secuencias finitas y genera un espectro discreto, mientras que la DTFT puede manejar secuencias infinitas y produce un espectro continuo.

Basado en la estrategia "divide y vencerás", descompone la DFT en subproblemas más pequeños.

calcular la transformada de Fourier $X(\omega)$ para cada una de las señales $x(t)$ dadas, sin usar propiedades de la transformada de Fourier. La definición de la transformada de Fourier para una señal continua en el tiempo $x(t)$ es:

$$ X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt $$

Donde:

$X(\omega)$ es la transformada de Fourier de $x(t)$.
$\omega$ es la frecuencia angular en radianes por segundo.
$j$ es la unidad imaginaria ($\sqrt{-1}$).
Aquí te presento cómo abordar cada inciso, aplicando la definición de la transformada de Fourier:

a) $x(t) = e^{-a|t|}, \quad a \in \mathbb{R}^{+}$

Necesitamos evaluar las dos integrales que establecimos: $$ X(\omega) = \int_{-\infty}^{0} e^{(a-j\omega)t} dt + \int_{0}^{\infty} e^{-(a+j\omega)t} dt $$

Para la primera integral: $$ \int_{-\infty}^{0} e^{(a-j\omega)t} dt = \left[ \frac{e^{(a-j\omega)t}}{a-j\omega} \right]_{-\infty}^{0} $$ Dado que $a > 0$, cuando $t \to -\infty$,  $e^{(a-j\omega)t} \to 0$. $$ = \frac{e^0}{a-j\omega} - 0 = \frac{1}{a-j\omega} $$

Para la segunda integral: $$ \int_{0}^{\infty} e^{-(a+j\omega)t} dt = \left[ \frac{e^{-(a+j\omega)t}}{-(a+j\omega)} \right]_{0}^{\infty} $$ Dado que $a > 0$, cuando $t \to \infty$, $e^{-(a+j\omega)t} \to 0$. $$ = 0 - \frac{e^0}{-(a+j\omega)} = \frac{1}{a+j\omega} $$

Sumando los dos resultados: $$ X(\omega) = \frac{1}{a-j\omega} + \frac{1}{a+j\omega} = \frac{a+j\omega + a-j\omega}{(a-j\omega)(a+j\omega)} = \frac{2a}{a^2 + \omega^2} $$

Entonces, la Transformada de Fourier de $e^{-a|t|}$ es $X(\omega) = \frac{2a}{a^2 + \omega^2}$.

b) $x(t) = \cos(w_c t), \quad w_c \in \mathbb{R}$

Tenemos: $$ X(\omega) = \frac{1}{2} \int_{-\infty}^{\infty} e^{j(w_c - \omega)t} dt + \frac{1}{2} \int_{-\infty}^{\infty} e^{-j(w_c + \omega)t} dt $$ Sabemos que $\int_{-\infty}^{\infty} e^{j\alpha t} dt = 2\pi \delta(\alpha)$

Entonces, la primera integral es: $\int_{-\infty}^{\infty} e^{j(w_c - \omega)t} dt = 2\pi \delta(w_c - \omega) = 2\pi \delta(\omega - w_c)$.

Y la segunda integral es $\int_{-\infty}^{\infty} e^{-j(w_c + \omega)t} dt = \int_{-\infty}^{\infty} e^{j(-(w_c + \omega))t} dt = 2\pi \delta(-(w_c + \omega)) = 2\pi \delta(w_c + \omega) = 2\pi \delta(\omega + w_c)$.

Sustituyendo de nuevo: $$ X(\omega) = \frac{1}{2} (2\pi \delta(\omega - w_c)) + \frac{1}{2} (2\pi \delta(\omega + w_c)) = \pi \delta(\omega - w_c) + \pi \delta(\omega + w_c) $$ Entonces, la Transformada de Fourier de $\cos(w_c t)$ es $X(\omega) = \pi (\delta(\omega - w_c) + \delta(\omega + w_c))$.

c) $x(t) = \sin(w_s t), \quad w_s \in \mathbb{R}$

Tenemos: $$ X(\omega) = \frac{1}{2j} \int_{-\infty}^{\infty} e^{j(w_s - \omega)t} dt - \frac{1}{2j} \int_{-\infty}^{\infty} e^{-j(w_s + \omega)t} dt $$ Usando la misma propiedad que en la parte b): La primera integral es $\int_{-\infty}^{\infty} e^{j(w_s - \omega)t} dt = 2\pi \delta(w_s - \omega) = 2\pi \delta(\omega - w_s)$. La segunda integral es $\int_{-\infty}^{\infty} e^{-j(w_s + \omega)t} dt = 2\pi \delta(\omega + w_s)$.

Sustituyendo de nuevo: $$ X(\omega) = \frac{1}{2j} (2\pi \delta(\omega - w_s)) - \frac{1}{2j} (2\pi \delta(\omega + w_s)) = \frac{\pi}{j} \delta(\omega - w_s) - \frac{\pi}{j} \delta(\omega + w_s) $$ Dado que $\frac{1}{j} = -j$: $$ X(\omega) = -j\pi \delta(\omega - w_s) + j\pi \delta(\omega + w_s) = j\pi (\delta(\omega + w_s) - \delta(\omega - w_s)) $$ Entonces, la Transformada de Fourier de $\sin(w_s t)$$\sin(w_s t)$ es $X(\omega) = j\pi (\delta(\omega + w_s) - \delta(\omega - w_s))$.

d) $x(t) = f(t)\cos(w_c t), \quad w_c \in \mathbb{R}, \quad f(t) \in \mathbb{R} \text{ o } \mathbb{C}$

Tenemos: $$ X(\omega) = \frac{1}{2} \int_{-\infty}^{\infty} f(t) e^{j(w_c - \omega)t} dt + \frac{1}{2} \int_{-\infty}^{\infty} f(t) e^{-j(w_c + \omega)t} dt $$ Sea $F(\omega)$ la Transformada de Fourier de $f(t)$: $F(\omega) = \int_{-\infty}^{\infty} f(t) e^{-j\omega t} dt$. La primera integral se puede reescribir como: $$ \int_{-\infty}^{\infty} f(t) e^{-j(\omega - w_c)t} dt $$ Esta es la Transformada de Fourier de $f(t)$ evaluada en la frecuencia $\omega - w_c$, que es $F(\omega - w_c)$.

La segunda integral se puede reescribir como: $$ \int_{-\infty}^{\infty} f(t) e^{-j(\omega + w_c)t} dt $$ Esta es la Transformada de Fourier de $f(t)$ evaluada en la frecuencia $\omega + w_c$$\omega + w_c$, que es $F(\omega + w_c)$.

Sustituyendo de nuevo: $$ X(\omega) = \frac{1}{2} F(\omega - w_c) + \frac{1}{2} F(\omega + w_c) $$ Entonces, la Transformada de Fourier de $f(t)\cos(w_c t)$ es $X(\omega) = \frac{1}{2} [F(\omega - w_c) + F(\omega + w_c)]$, donde$F(\omega)$ es la Transformada de Fourier de $f(t)$.

e) $x(t) = e^{-a|t|^2}, \quad a \in \mathbb{R}^{+}$

Necesitamos evaluar: $$ X(\omega) = \int_{-\infty}^{\infty} e^{-at^2} e^{-j\omega t} dt = \int_{-\infty}^{\infty} e^{-at^2 - j\omega t} dt $$ Para resolver esto usando la fórmula de la integral Gaussiana, completamos el cuadrado en el exponente: $$ -at^2 - j\omega t = -a \left( t^2 + \frac{j\omega}{a} t \right) $$ Para completar el cuadrado de $t^2 + \frac{j\omega}{a} t$, sumamos y restamos $(\frac{j\omega}{2a})^2$:
$$ t^2 + \frac{j\omega}{a} t + \left(\frac{j\omega}{2a}\right)^2 - \left(\frac{j\omega}{2a}\right)^2 = \left( t + \frac{j\omega}{2a} \right)^2 - \frac{(j\omega)^2}{4a^2} = \left( t + \frac{j\omega}{2a} \right)^2 + \frac{\omega^2}{4a^2} $$ Entonces el exponente se convierte en: $$ -a \left[ \left( t + \frac{j\omega}{2a} \right)^2 + \frac{\omega^2}{4a^2} \right] = -a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{a\omega^2}{4a^2} = -a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a} $$ Ahora la integral es: $$ X(\omega) = \int_{-\infty}^{\infty} e^{-a \left( t + \frac{j\omega}{2a} \right)^2 - \frac{\omega^2}{4a}} dt = e^{-\frac{\omega^2}{4a}} \int_{-\infty}^{\infty} e^{-a \left( t + \frac{j\omega}{2a} \right)^2} dt $$ Sea $\tau = t + \frac{j\omega}{2a}$. Entonces $d\tau = dt$. Los límites de integración permanecen de $-\infty$ a $\infty$ en el plano complejo. $$ X(\omega) = e^{-\frac{\omega^2}{4a}} \int_{-\infty}^{\infty} e^{-a \tau^2} d\tau $$ Esta es una integral Gaussiana estándar con $b=a$. $$ \int_{-\infty}^{\infty} e^{-a \tau^2} d\tau = \sqrt{\frac{\pi}{a}} $$ Entonces, $$ X(\omega) = e^{-\frac{\omega^2}{4a}} \sqrt{\frac{\pi}{a}} = \sqrt{\frac{\pi}{a}} e^{-\frac{\omega^2}{4a}} $$ Entonces, la Transformada de Fourier de $e^{-at^2}$ es $X(\omega) = \sqrt{\frac{\pi}{a}} e^{-\frac{\omega^2}{4a}}$.

f) $x(t) = A \cdot \text{Arct}_d(t), \quad A, d \in \mathbb{R}$

calcular la transformada de Fourier de $x(t) = A \cdot \text{Arct}_d(t)$ utilizando la definición:

$$ X(\omega) = \int_{-\infty}^{\infty} x(t) e^{-j\omega t} dt $$

La función $\text{Arct}_d(t)$ se refiere a la función rectangular, definida como:

$$ \text{rect}\left(\frac{t}{d}\right) = \begin{cases} 1 & \text{si } |t| \le \frac{d}{2} \\ 0 & \text{si } |t| > \frac{d}{2} \end{cases} $$

Por lo tanto, la señal es:

$$ x(t) = A \cdot \text{rect}\left(\frac{t}{d}\right) = \begin{cases} A & \text{si } -\frac{d}{2} \le t \le \frac{d}{2} \\ 0 & \text{en otro caso} \end{cases} $$

Aplicando la definición de la transformada de Fourier e integrando sobre el intervalo donde la señal es no nula:

$$ X(\omega) = \int_{-d/2}^{d/2} A e^{-j\omega t} dt $$ $$ X(\omega) = A \int_{-d/2}^{d/2} e^{-j\omega t} dt $$

Para $\omega \neq 0$:

$$ X(\omega) = A \left[ \frac{e^{-j\omega t}}{-j\omega} \right]_{-d/2}^{d/2} $$ $$ X(\omega) = A \left( \frac{e^{-j\omega d/2} - e^{j\omega d/2}}{-j\omega} \right) $$ $$ X(\omega) = A \left( \frac{-(e^{j\omega d/2} - e^{-j\omega d/2})}{-j\omega} \right) $$

Usando la identidad $\sin(\theta) = \frac{e^{j\theta} - e^{-j\theta}}{2j}$:

$$ X(\omega) = A \left( \frac{-2j \sin(\omega d/2)}{-j\omega} \right) $$ $$ X(\omega) = A \left( \frac{2 \sin(\omega d/2)}{\omega} \right) $$ $$ X(\omega) = Ad \left( \frac{\sin(\omega d/2)}{\omega d/2} \right) $$

Para $\omega = 0$:

$$ X(0) = \int_{-d/2}^{d/2} A dt = A [t]_{-d/2}^{d/2} = Ad $$

La expresión $Ad \left( \frac{\sin(\omega d/2)}{\omega d/2} \right)$ es la función sinc no normalizada multiplicada por $Ad$.

La transformada de Fourier de $x(t) = A \cdot \text{rect}\left(\frac{t}{d}\right)$ es:

$$ X(\omega) = Ad \frac{\sin(\frac{\omega d}{2})}{\frac{\omega d}{2}} $$

La modulación en amplitud (AM) es un método que permite enviar información, como señales de audio, mediante la variación de la amplitud de una onda portadora de frecuencia más alta. En la detección coherente —también llamada síncrona— el receptor utiliza una versión idéntica de la portadora original, con igual frecuencia y fase, para extraer la señal transmitida.

Este proceso consiste en multiplicar la señal AM que se recibe por una señal portadora generada internamente en el receptor. Luego, se aplica un filtro pasabajo para eliminar las frecuencias altas resultantes, dejando solamente la señal original que se deseaba recuperar.

Una de las principales fortalezas de la detección coherente es su alta tolerancia al ruido y a las interferencias. Dado que el receptor está sincronizado con la portadora transmitida, puede diferenciar con gran precisión la señal útil de otras señales no deseadas. No obstante, su inconveniente principal radica en la complejidad técnica y el costo asociados con la generación de una portadora local que esté perfectamente sincronizada.

Ejemplo en Python
Se puede implementar una simulación usando scipy.fft.rfft para analizar el espectro de señales AM con mensajes como pulsos rectangulares o cosenos, demostrando la demodulación coherente.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.signal import butter, lfilter

# --- Parámetros ---
Fs = 10000  # Frecuencia de muestreo
T = 1/Fs    # Período de muestreo
tiempo = 1  # Duración de la simulación en segundos
t = np.arange(0, tiempo, T) # Vector de tiempo

Fc = 1000 # Frecuencia de la portadora (debe ser mucho mayor que la frecuencia del mensaje)

# Índice de modulación (mu)
mu = 0.8

# --- Diseño del Filtro Pasa-Bajos (para demodulación) ---
orden = 5
frecuencia_corte = 2 * Fc / Fs # Frecuencia de corte normalizada
b, a = butter(orden, frecuencia_corte, btype='low')

# --- Función para Modulación AM ---
def modular_am(mensaje, Ac, Fc, Fs, mu):
    t = np.arange(0, len(mensaje)/Fs, 1/Fs)
    portadora = Ac * np.cos(2 * np.pi * Fc * t)
    senal_modulada = Ac * (1 + mu * mensaje) * np.cos(2 * np.pi * Fc * t)
    return senal_modulada, portadora

# --- Función para Demodulación Coherente ---
def demodular_am_coherente(senal_recibida, Fc, Fs, b, a):
    t = np.arange(0, len(senal_recibida)/Fs, 1/Fs)
    portadora_local = 2 * np.cos(2 * np.pi * Fc * t) # Se multiplica por 2 para compensar el factor 1/2 en la demodulación ideal
    senal_multiplicada = senal_recibida * portadora_local
    senal_demodulada_filtrada = lfilter(b, a, senal_multiplicada)
    # Eliminar componente DC para recuperar el mensaje (aproximado)
    senal_demodulada = (senal_demodulada_filtrada / Ac) - 1
    return senal_demodulada

# --- Parámetros de la señal mensaje ---
Am = 0.5 # Amplitud de la señal mensaje
Ac = 1   # Amplitud de la portadora

# --- Caso 1: Señal Mensaje tipo Pulso Rectangular ---
duracion_pulso = 0.1
pulso = np.zeros_like(t)
pulso[int(0.4*Fs):int(0.6*Fs)] = Am
mensaje_pulso = pulso

# Modulación
senal_modulada_pulso, portadora_pulso = modular_am(mensaje_pulso, Ac, Fc, Fs, mu)

# Demodulación
senal_demodulada_pulso = demodular_am_coherente(senal_modulada_pulso, Fc, Fs, b, a)

# --- Gráficas en el Dominio del Tiempo (Pulso Rectangular) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_pulso)
plt.title('Señal Mensaje (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_pulso)
plt.title('Señal Modulada AM (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_pulso)
plt.title('Señal Demodulada AM Coherente (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la región del pulso

plt.tight_layout()
plt.show()

# --- Gráficas en el Dominio de la Frecuencia (Pulso Rectangular) ---
# Calcular la transformada de Fourier real
mensaje_pulso_fft = rfft(mensaje_pulso)
senal_modulada_pulso_fft = rfft(senal_modulada_pulso)
senal_demodulada_pulso_fft = rfft(senal_demodulada_pulso)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_pulso_fft))
plt.title('Espectro de la Señal Mensaje (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_pulso_fft))
plt.title('Espectro de la Señal Modulada AM (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de interés

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_pulso_fft))
plt.title('Espectro de la Señal Demodulada AM Coherente (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.tight_layout()
plt.show()

# --- Caso 2: Señal Mensaje tipo Coseno ---
Fm = 10  # Frecuencia del mensaje
mensaje_coseno = Am * np.cos(2 * np.pi * Fm * t)

# Modulación
senal_modulada_coseno, portadora_coseno = modular_am(mensaje_coseno, Ac, Fc, Fs, mu)

# Demodulación
senal_demodulada_coseno = demodular_am_coherente(senal_modulada_coseno, Fc, Fs, b, a)

# --- Gráficas en el Dominio del Tiempo (Coseno) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_coseno)
plt.title('Señal Mensaje (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_coseno)
plt.title('Señal Modulada AM (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_coseno)
plt.title('Señal Demodulada AM Coherente (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.tight_layout()
plt.show()

# --- Gráficas en el Dominio de la Frecuencia (Coseno) ---
# Calcular la transformada de Fourier real
mensaje_coseno_fft = rfft(mensaje_coseno)
senal_modulada_coseno_fft = rfft(senal_modulada_coseno)
senal_demodulada_coseno_fft = rfft(senal_demodulada_coseno)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_coseno_fft))
plt.title('Espectro de la Señal Mensaje (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_coseno_fft))
plt.title('Espectro de la Señal Modulada AM (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de interés

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_coseno_fft))
plt.title('Espectro de la Señal Demodulada AM Coherente (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de interés

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
from IPython.display import Audio

# Cargar audio
# Make sure to upload 'fragmento_cancion.wav' to your Colab environment
fs, audio = wavfile.read('/content/sample_data/fragmento_cancion.wav')  # asegúrate que es MONO

if audio.ndim == 2:
    audio = audio[:, 0]

audio = audio / np.max(np.abs(audio))  # Normalizar

# Use the actual duration of the audio
dur = len(audio) / fs
N = len(audio)
t = np.linspace(0, dur, N, endpoint=False)
mensaje = audio

# Parámetros de modulación
fc = 10000  # Frecuencia portadora en Hz
Ac = 1      # Amplitud portadora
m_index = 1 # Índice de modulación

# Portadora
portadora = Ac * np.cos(2 * np.pi * fc * t)

# Señal AM
s_am = Ac * (1 + m_index * mensaje) * np.cos(2 * np.pi * fc * t)

# Espectros
f = rfftfreq(N, 1/fs)
M_f = np.abs(rfft(mensaje))
P_f = np.abs(rfft(portadora))
AM_f = np.abs(rfft(s_am))

# Graficar señales en el tiempo
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(t, mensaje)
plt.title('Mensaje en el tiempo')
plt.xlabel('Tiempo [s]')

plt.subplot(3, 2, 2)
plt.plot(f, M_f)
plt.title('Espectro del mensaje')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2]) # Limit frequency axis for better visualization


plt.subplot(3, 2, 3)
plt.plot(t, portadora)
plt.title('Portadora en el tiempo')
plt.xlabel('Tiempo [s]')


plt.subplot(3, 2, 4)
plt.plot(f, P_f)
plt.title('Espectro de la portadora')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2]) # Limit frequency axis for better visualization


plt.subplot(3, 2, 5)
plt.plot(t, s_am)
plt.title('Señal AM en el tiempo')
plt.xlabel('Tiempo [s]')


plt.subplot(3, 2, 6)
plt.plot(f, AM_f)
plt.title('Espectro de la señal AM')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2 + 200]) # Limit frequency axis for better visualization


plt.tight_layout()
plt.show()

# Reproducción de audio (mensaje, portadora, señal AM)
print("Reproduciendo señal mensaje (audio original):")
display(Audio(mensaje, rate=fs))

print("Reproduciendo portadora (sonido puro):")
display(Audio(portadora, rate=fs))

print("Reproduciendo señal AM:")
display(Audio(s_am, rate=fs))
```

5. Conclusiones
El THD es un indicador clave de la calidad de la energía en sistemas con cargas no lineales.

El FP se degrada con altos niveles de THD, especialmente en rectificadores con filtro capacitivo.

Para mejorar el FP, se pueden usar filtros pasivos o activos para reducir los armónicos.
