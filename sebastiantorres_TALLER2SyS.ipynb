3. **Convoluci√≥n temporal:** $f_1(t)*f_2(t)$ se eval√∫a mediante integral:
   $
   y(t) = \int_{0}^{t} \Bigl( \frac{7}{12} e^{-6\tau} e^{-3j\tau} \Bigr) \cdot \Bigl( 10 (t-\tau) e^{-24(t-\tau)} \Bigr) \, d\tau.
   $

---

### d) $\mathcal{F} \{ 3t^3 \}$

1. **Derivada en frecuencia:** propiedad $t^n x(t) ‚Üî{\mathcal{F}} j^n \frac{d^n}{d\omega^n} X(\omega)$.
2. **Caso base:** $1 ‚Üî{\mathcal{F}} 2\pi\delta(\omega)$. Luego,
   $
   \mathcal{F}\{3t^3\} = 3j^3 2\pi\delta^{(3)}(\omega) = -6\pi j \, \delta^{(3)}(\omega).
   $

Ejemplo en Python
Se puede implementar una simulaci√≥n usando scipy.fft.rfft para analizar el espectro de se√±ales AM con mensajes como pulsos rectangulares o cosenos, demostrando la demodulaci√≥n coherente.



```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import rfft, rfftfreq
from scipy.signal import butter, lfilter

# --- Par√°metros ---
Fs = 10000  # Frecuencia de muestreo
T = 1/Fs    # Per√≠odo de muestreo
tiempo = 1  # Duraci√≥n de la simulaci√≥n en segundos
t = np.arange(0, tiempo, T) # Vector de tiempo

Fc = 1000 # Frecuencia de la portadora (debe ser mucho mayor que la frecuencia del mensaje)

# √çndice de modulaci√≥n (mu)
mu = 0.8

# --- Dise√±o del Filtro Pasa-Bajos (para demodulaci√≥n) ---
orden = 5
frecuencia_corte = 2 * Fc / Fs # Frecuencia de corte normalizada
b, a = butter(orden, frecuencia_corte, btype='low')

# --- Funci√≥n para Modulaci√≥n AM ---
def modular_am(mensaje, Ac, Fc, Fs, mu):
    t = np.arange(0, len(mensaje)/Fs, 1/Fs)
    portadora = Ac * np.cos(2 * np.pi * Fc * t)
    senal_modulada = Ac * (1 + mu * mensaje) * np.cos(2 * np.pi * Fc * t)
    return senal_modulada, portadora

# --- Funci√≥n para Demodulaci√≥n Coherente ---
def demodular_am_coherente(senal_recibida, Fc, Fs, b, a):
    t = np.arange(0, len(senal_recibida)/Fs, 1/Fs)
    portadora_local = 2 * np.cos(2 * np.pi * Fc * t) # Se multiplica por 2 para compensar el factor 1/2 en la demodulaci√≥n ideal
    senal_multiplicada = senal_recibida * portadora_local
    senal_demodulada_filtrada = lfilter(b, a, senal_multiplicada)
    # Eliminar componente DC para recuperar el mensaje (aproximado)
    senal_demodulada = (senal_demodulada_filtrada / Ac) - 1
    return senal_demodulada

# --- Par√°metros de la se√±al mensaje ---
Am = 0.5 # Amplitud de la se√±al mensaje
Ac = 1   # Amplitud de la portadora

# --- Caso 1: Se√±al Mensaje tipo Pulso Rectangular ---
duracion_pulso = 0.1
pulso = np.zeros_like(t)
pulso[int(0.4*Fs):int(0.6*Fs)] = Am
mensaje_pulso = pulso

# Modulaci√≥n
senal_modulada_pulso, portadora_pulso = modular_am(mensaje_pulso, Ac, Fc, Fs, mu)

# Demodulaci√≥n
senal_demodulada_pulso = demodular_am_coherente(senal_modulada_pulso, Fc, Fs, b, a)

# --- Gr√°ficas en el Dominio del Tiempo (Pulso Rectangular) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_pulso)
plt.title('Se√±al Mensaje (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la regi√≥n del pulso

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_pulso)
plt.title('Se√±al Modulada AM (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la regi√≥n del pulso

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_pulso)
plt.title('Se√±al Demodulada AM Coherente (Pulso Rectangular) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0.3, 0.7]) # Enfocar en la regi√≥n del pulso

plt.tight_layout()
plt.show()

# --- Gr√°ficas en el Dominio de la Frecuencia (Pulso Rectangular) ---
# Calcular la transformada de Fourier real
mensaje_pulso_fft = rfft(mensaje_pulso)
senal_modulada_pulso_fft = rfft(senal_modulada_pulso)
senal_demodulada_pulso_fft = rfft(senal_demodulada_pulso)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_pulso_fft))
plt.title('Espectro de la Se√±al Mensaje (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de inter√©s

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_pulso_fft))
plt.title('Espectro de la Se√±al Modulada AM (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de inter√©s

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_pulso_fft))
plt.title('Espectro de la Se√±al Demodulada AM Coherente (Pulso Rectangular)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de inter√©s

plt.tight_layout()
plt.show()

# --- Caso 2: Se√±al Mensaje tipo Coseno ---
Fm = 10  # Frecuencia del mensaje
mensaje_coseno = Am * np.cos(2 * np.pi * Fm * t)

# Modulaci√≥n
senal_modulada_coseno, portadora_coseno = modular_am(mensaje_coseno, Ac, Fc, Fs, mu)

# Demodulaci√≥n
senal_demodulada_coseno = demodular_am_coherente(senal_modulada_coseno, Fc, Fs, b, a)

# --- Gr√°ficas en el Dominio del Tiempo (Coseno) ---
plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(t, mensaje_coseno)
plt.title('Se√±al Mensaje (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 2)
plt.plot(t, senal_modulada_coseno)
plt.title('Se√±al Modulada AM (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.subplot(3, 1, 3)
plt.plot(t, senal_demodulada_coseno)
plt.title('Se√±al Demodulada AM Coherente (Coseno) en el Tiempo')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.grid(True)
plt.xlim([0, 0.5]) # Mostrar varias oscilaciones del mensaje

plt.tight_layout()
plt.show()

# --- Gr√°ficas en el Dominio de la Frecuencia (Coseno) ---
# Calcular la transformada de Fourier real
mensaje_coseno_fft = rfft(mensaje_coseno)
senal_modulada_coseno_fft = rfft(senal_modulada_coseno)
senal_demodulada_coseno_fft = rfft(senal_demodulada_coseno)

# Obtener las frecuencias correspondientes
frecuencias = rfftfreq(len(t), 1/Fs)

plt.figure(figsize=(12, 8))

plt.subplot(3, 1, 1)
plt.plot(frecuencias, np.abs(mensaje_coseno_fft))
plt.title('Espectro de la Se√±al Mensaje (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de inter√©s

plt.subplot(3, 1, 2)
plt.plot(frecuencias, np.abs(senal_modulada_coseno_fft))
plt.title('Espectro de la Se√±al Modulada AM (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2 + 200]) # Rango de frecuencias de inter√©s

plt.subplot(3, 1, 3)
plt.plot(frecuencias, np.abs(senal_demodulada_coseno_fft))
plt.title('Espectro de la Se√±al Demodulada AM Coherente (Coseno)')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.grid(True)
plt.xlim([0, Fc * 2]) # Rango de frecuencias de inter√©s

plt.tight_layout()
plt.show()
```

Aplicaci√≥n en comunicaciones - modulaci√≥n AM

Sea la se√±al portadora $c(t) = A_c \cos(2\pi F_c t)$, con $A_c, F_c \in \mathbb{R}$, y la se√±al mensaje $m(t) \in \mathbb{R}$.

Encuentre el espectro en frecuencia de la se√±al modulada en amplitud (AM):

$$
y(t) = \left(1 + \frac{m(t)}{A_c} \right)c(t)
$$

Luego, descargue desde YouTube, 5 segundos de su canci√≥n favorita (capturando del segundo 20 al 25).

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from scipy.fft import rfft, rfftfreq
from IPython.display import Audio

# Cargar audio
# Make sure to upload 'fragmento_cancion.wav' to your Colab environment
fs, audio = wavfile.read('/content/sample_data/fragmento_cancion.wav')  # aseg√∫rate que es MONO

if audio.ndim == 2:
    audio = audio[:, 0]

audio = audio / np.max(np.abs(audio))  # Normalizar

# Use the actual duration of the audio
dur = len(audio) / fs
N = len(audio)
t = np.linspace(0, dur, N, endpoint=False)
mensaje = audio

# Par√°metros de modulaci√≥n
fc = 10000  # Frecuencia portadora en Hz
Ac = 1      # Amplitud portadora
m_index = 1 # √çndice de modulaci√≥n

# Portadora
portadora = Ac * np.cos(2 * np.pi * fc * t)

# Se√±al AM
s_am = Ac * (1 + m_index * mensaje) * np.cos(2 * np.pi * fc * t)

# Espectros
f = rfftfreq(N, 1/fs)
M_f = np.abs(rfft(mensaje))
P_f = np.abs(rfft(portadora))
AM_f = np.abs(rfft(s_am))

# Graficar se√±ales en el tiempo
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(t, mensaje)
plt.title('Mensaje en el tiempo')
plt.xlabel('Tiempo [s]')

plt.subplot(3, 2, 2)
plt.plot(f, M_f)
plt.title('Espectro del mensaje')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2]) # Limit frequency axis for better visualization


plt.subplot(3, 2, 3)
plt.plot(t, portadora)
plt.title('Portadora en el tiempo')
plt.xlabel('Tiempo [s]')


plt.subplot(3, 2, 4)
plt.plot(f, P_f)
plt.title('Espectro de la portadora')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2]) # Limit frequency axis for better visualization


plt.subplot(3, 2, 5)
plt.plot(t, s_am)
plt.title('Se√±al AM en el tiempo')
plt.xlabel('Tiempo [s]')


plt.subplot(3, 2, 6)
plt.plot(f, AM_f)
plt.title('Espectro de la se√±al AM')
plt.xlabel('Frecuencia [Hz]')
plt.xlim([0, fc * 2 + 200]) # Limit frequency axis for better visualization


plt.tight_layout()
plt.show()

# Reproducci√≥n de audio (mensaje, portadora, se√±al AM)
print("Reproduciendo se√±al mensaje (audio original):")
display(Audio(mensaje, rate=fs))

print("Reproduciendo portadora (sonido puro):")
display(Audio(portadora, rate=fs))

print("Reproduciendo se√±al AM:")
display(Audio(s_am, rate=fs))
```
